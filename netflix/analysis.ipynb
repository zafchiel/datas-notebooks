```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix
```

```python
netflix.info()
```

```python
netflix.dropna(inplace=True)
netflix
```

```python
netflix.reset_index(drop=True)
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
```

```python
netflix.info()
```

```python
types = netflix['type'].value_counts()
types
```

```python
types.plot.bar()
plt.title('Types of shows')
plt.ylabel('Count of shows')

plt.show()
```

```python
netflix['director'].value_counts().sort_values(ascending=False).iloc[0:10]
```

```python
netflix.groupby(['director', 'type'])['director'].value_counts().sort_values(ascending=False).iloc[0:10]
```

```python
netflix.groupby(['country', 'type'])['country'].value_counts().sort_values(ascending=False).iloc[0:10]
```

```python
netflix[netflix['type'] == 'TV Show'].groupby(['country', 'type'])['country'].value_counts().sort_values(ascending=False).iloc[0:10]
```

```python
netflix[netflix['type'] == 'Movie'].groupby('release_year')['release_year'].value_counts().sort_values(ascending=False).iloc[0:10]
```

```python
netflix['release_year'].plot.hist(bins=40)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
netflix['rating'].value_counts()
```

```python
durations = netflix.groupby(['duration', 'type'])['duration'].value_counts().reset_index()
durations
```

```python
movies = netflix[netflix['type'] == 'Movie']
movies.loc[:, 'duration'] = movies['duration'].str[:-4].astype(int)

duration_by_year = movies.groupby('release_year')['duration'].mean()
duration_by_year.plot(figsize=(20,6))
plt.title('Average duration of films by year')
plt.ylabel('Average duration in minutes')
plt.xlabel('Release year')

plt.show()
```

```python
movie_durations = durations[durations['type'] == 'Movie'].sort_values(by='count', ascending=False).iloc[0:10]
movie_durations
```

```python
show_durations = durations[durations['type'] == 'TV Show'].sort_values(by='count', ascending=False).iloc[0:10]
show_durations
```
