```python
positions = df['Position'].value_counts()
positions
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Positions')
plt.ylabel('Number of players')

# for i in range(len(positions)):
#     plt.text(x=i, y=positions.iloc[i], s=positions.iloc[i], ha='center', va='bottom')

bars = plt.bar(positions.index, positions.array, edgecolor='white', linewidth=1, width=0.5)

plt.bar_label(bars)

plt.show()
```

```python
goals = df.groupby('Club')['Goals'].sum().sort_values(ascending=False).reset_index()
goals
```

```python
plt.figure(figsize=(10, 5))

plt.xlabel('Clubs')
plt.ylabel('Goals')

plt.xticks(rotation=90)

bars = plt.bar(goals['Club'], goals['Goals'], width=0.7)
plt.bar_label(bars, padding=-15)

plt.show()
```

```python
goals_and_assists = df.groupby('Club').agg({
    'Goals': 'sum',
    'Assists': 'sum'
}).sort_values(by=['Goals'], ascending=False)
goals_and_assists
```

```python
plt.figure(figsize=(20,6))

plt.xticks(rotation=90)
plt.xlabel('Clubs')

bar_width = 0.5
x = np.arange(len(goals_and_assists))

bars1 = plt.bar(goals_and_assists.index, goals_and_assists['Goals'], width=bar_width)
bars2 = plt.bar(x + bar_width, goals_and_assists['Assists'], width=bar_width)

plt.bar_label(bars1)
plt.bar_label(bars2)
plt.legend(['Goals', 'Assists'], loc='upper right')

plt.show()
```
