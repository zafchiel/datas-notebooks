```python
positions = df['Position'].value_counts()
positions
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Positions')
plt.ylabel('Number of players')

# for i in range(len(positions)):
#     plt.text(x=i, y=positions.iloc[i], s=positions.iloc[i], ha='center', va='bottom')

bars = plt.bar(positions.index, positions.array, edgecolor='white', linewidth=1, width=0.5)

plt.bar_label(bars)

plt.show()
```

```python
goals = df.groupby('Club')['Goals'].sum().sort_values(ascending=False).reset_index()
goals
```

```python
plt.figure(figsize=(10, 5))

plt.xlabel('Clubs')
plt.ylabel('Goals')

plt.xticks(rotation=90)

bars = plt.bar(goals['Club'], goals['Goals'], width=0.7)
plt.bar_label(bars, padding=-15)

plt.show()
```

```python
goals_and_assists = df.groupby('Club').agg({
    'Goals': 'sum',
    'Assists': 'sum'
}).sort_values(by=['Goals'], ascending=False)
goals_and_assists
```

```python
plt.figure(figsize=(20,6))
plt.title('Total Goals and Assist by Club')

plt.xticks(rotation=90)
plt.xlabel('Clubs')

bar_width = 0.5
x = np.arange(len(goals_and_assists))

bars1 = plt.bar(goals_and_assists.index, goals_and_assists['Goals'], width=bar_width)
bars2 = plt.bar(x + bar_width, goals_and_assists['Assists'], width=bar_width)

plt.bar_label(bars1)
plt.bar_label(bars2)
plt.legend(['Goals', 'Assists'], loc='upper right')

plt.show()
```

```python
red_yellow = df.groupby('Club')[['Yellow_Cards', 'Red_Cards']].sum()
red_yellow
```

```python
chart = red_yellow.plot.bar(figsize=(10,3), color=['yellow', 'red'])
chart.margins(0.2)

for container in chart.containers:
    chart.bar_label(container)
```

```python
# Top 5 scorers
df.loc[:, ['Name', 'Club', 'Goals']].sort_values(by=['Goals'], ascending=False).head(5).reset_index(drop=True)
```

```python
top_goals = df['Goals'].sort_values(ascending=False).head(5).values

df[df['Goals'].isin(top_goals)].loc[:, ['Name', 'Club', 'Goals']].sort_values(by=['Goals'], ascending=False).reset_index(drop=True)
```

```python
penalty_not_scored = df['Penalty_Attempted'].sum() - df['Penalty_Goals'].sum()

data = [penalty_not_scored, df['Penalty_Goals'].sum()]
labels = ['Penalties missed', 'Penalties scored']

plt.title('Penalties Missed vs Penalties Scored')

plt.pie(data, labels=labels, autopct='%.1f%%', explode=[0.1, 0.1])
```

```python
avg_age = df.groupby('Club')['Age'].mean().sort_values()

plt.figure(figsize=(6,13))
plt.margins(0.2, 0.01)

hbars = plt.barh(avg_age.index, avg_age.values, color='#56ad74')
plt.title('Average age in club')

plt.bar_label(hbars, padding=2)

plt.show()
```

```python
Chelsea = df[df['Club'] == 'Chelsea']
Chelsea
```

```python
Chelsea['Name']
```

```python
nationality = Chelsea['Nationality'].value_counts()

plt.pie(nationality.values, labels=nationality.index, autopct='%.1f%%', pctdistance=1.4, labeldistance=1.05)
plt.title('Chelsea\'s players nationality')
plt.show()
```

```python
# Average age by position
avg_age_by_position = Chelsea.groupby('Position')['Age'].mean()
avg_age_by_position
```

```python
bars = plt.bar(avg_age_by_position.index, avg_age_by_position.values.round())
plt.bar_label(bars)
plt.xlabel('Position')
plt.ylabel('Average height')
plt.title('Average age by position in Chelsea')
```
